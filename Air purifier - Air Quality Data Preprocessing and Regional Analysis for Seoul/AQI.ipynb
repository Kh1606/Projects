```python
import os
import pandas as pd

# Define the input folder containing the original CSV files
input_folder_path = '/Users/saidev/Downloads/aqii5/'  # Replace with the path to your folder with CSV files

# Define the output folder where the new combined CSV files will be saved
output_folder_path = '/Users/saidev/Downloads/aqii6/'  # Replace with the path to your output folder

# Ensure the output directory exists
os.makedirs(output_folder_path, exist_ok=True)

# Dictionary to hold lists of DataFrames for each prefix
combined_data = {}

# Loop over each CSV file in the input folder
for filename in os.listdir(input_folder_path):
    if filename.endswith('.csv'):
        # Extract the prefix (the first word) from the filename
        prefix = filename.split('_')[0]
        
        # Construct the full path to the CSV file
        file_path = os.path.join(input_folder_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Add the DataFrame to the corresponding prefix entry in the dictionary
        if prefix in combined_data:
            combined_data[prefix].append(df)
        else:
            combined_data[prefix] = [df]

# Combine DataFrames with the same prefix and save them to new CSV files
for prefix, dataframes in combined_data.items():
    # Combine all DataFrames for the current prefix
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    # Ensure the 'date' column is treated as a datetime for sorting
    combined_df.iloc[:, 0] = pd.to_datetime(combined_df.iloc[:, 0], format='%Y%m%d%H%M')
    
    # Sort by the 'date' column to ensure continuous date range
    combined_df = combined_df.sort_values(by=combined_df.columns[0])
    
    # Convert the 'date' column back to string format for saving
    combined_df.iloc[:, 0] = combined_df.iloc[:, 0].dt.strftime('%Y%m%d%H%M')
    
    # Save the combined DataFrame to a new CSV file
    output_file_path = os.path.join(output_folder_path, f"{prefix}_combined.csv")
    combined_df.to_csv(output_file_path, index=False)
    print(f"Created combined file: {output_file_path}")
```
